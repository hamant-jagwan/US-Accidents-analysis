###Start time

```python
us_df.Start_Time = pd.to_datetime(us_df.Start_Time) #converting the string in datetime formate
us_df.Start_Time[0]
```

```python
us_df.Start_Time.dt.hour #to get the hour from the datetime formate
sns.distplot(us_df.Start_Time.dt.hour , bins = 24 , norm_hist = True , kde = False)# plotting the hours of Start_Time column
```

- A high percentage of high accidents occurs in 4-6 pm and little bit in 6-8 probably this shows the working hours and the closing time of working.

```python
import matplotlib.pyplot as plt

us_df.Start_Time.dt.day_of_week # getting the day of the week in which accidents are happend
plt.hist(us_df.Start_Time.dt.day_of_week , bins = 7,  density = True) #plotting histogram to show the week days accidents
```

is distribution of accidents by hour the same on weekends as on week day


```python
sunday_start_time = us_df.Start_Time[us_df.Start_Time.dt.dayofweek == 6]#knowing the weekend accident distribution
sns.distplot(sunday_start_time.dt.hour , bins = 24 , norm_hist = True , kde = False)
```

- major accident in weekend is between 10:00am - 10:00pm and 12:00am- 3:00pm

```python
monday_start_time = us_df.Start_Time[us_df.Start_Time.dt.dayofweek == 0]#knowing the weekday(monday) accident distribution
sns.distplot(monday_start_time.dt.hour ,bins = 24, norm_hist = True, kde =  False)
```

in weekdays major accidents happen in 3:00pm - 7:00pm and little bit in 7:00am- 10:00 am

```python
sns.distplot(us_df.Start_Time.dt.month ,bins = 12, norm_hist = True, kde =  False)#getting the months which have highest numbers of accident 
```

december has the highest number of accident
